```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print ("libraries imported succesfully!")
```

```python
df = pd.read_csv("E-commerce customer Behavior - Sheet1.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df["Satisfaction Level"] = df["Satisfaction Level"].fillna(df["Satisfaction Level"].mode()[0])
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"],bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customer")
plt.show()
```

```python
df["Membership Type"].value_counts().plot(kind="bar")
plt.title("Membership Type Distributions")
plt.xlabel("Membership Type")
plt.ylabel("Number of Customers")
plt.show()
```

```python
df["City"].value_counts().plot(kind="bar",figsize=(10,5))
plt.title("Customers by city")
plt.xlabel("city")
plt.ylabel("Number of customers")
plt.xticks(rotation=45)
plt.show()
```

```python
membership_spend = df.groupby("Membership Type")["Total Spend"].mean()

plt.figure(figsize=(6,4))
membership_spend.plot(kind="bar", color="green")
plt.title("Average Spend by Membership Type")
plt.xlabel("Membership Type")
plt.ylabel("Average Total Spend")
plt.show()
```

```python
gender_spend = df.groupby("Gender")["Total Spend"].mean()

plt.figure(figsize=(5,4))
gender_spend.plot(kind="bar", color=["skyblue","pink"])
plt.title("Average Spend by Gender")
plt.ylabel("Average Total Spend")
plt.show()
```

```python
city_spend = df.groupby("City")["Total Spend"].mean().sort_values()

plt.figure(figsize=(8,5))
city_spend.plot(kind="barh")
plt.title("Average Spend by City")
plt.xlabel("Average Total Spend")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.to_csv("cleaned_ecommerce_data.csv", index=False)
print("Dataset saved successfully!")
```

```python
city_spend = df.groupby("City")["Total Spend"].sum().sort_values(ascending=False)
print(city_spend)

city_spend.plot(kind="bar", figsize=(8,5))
plt.title("Total Spend by City")
plt.ylabel("Total Spend")
plt.show()
```

```python
membership_spend = df.groupby("Membership Type")["Total Spend"].mean()
print(membership_spend)

membership_spend.plot(kind="bar", figsize=(6,4))
plt.title("Average Spending by Membership Type")
plt.ylabel("Average Spend")
plt.show()
```

```python
df.groupby("Age")["Total Spend"].mean().plot(figsize=(10,5))
plt.title("Average Spending by Age")
plt.ylabel("Average Spend")
plt.show()
```

```python
plt.scatter(df["Average Rating"], df["Total Spend"])
plt.xlabel("Average Rating")
plt.ylabel("Total Spend")
plt.title("Rating vs Total Spend")
plt.show()
```
